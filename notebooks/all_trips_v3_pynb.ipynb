```python
import pandas as pd

# Load the CSV into a DataFrame
df = pd.read_csv("all_trips_v3.csv")

# Preview the data
df.head()
```

```python
df.tail()
```

```python
counts = df['year'].value_counts()
print(counts)
```

```python
df.info()
```

```python
# Convert 'started_at' to datetime
df['started_at'] = pd.to_datetime(df['started_at'], format='mixed')

# Extract components from datetime
df['date'] = df['started_at'].dt.date
df['month'] = df['started_at'].dt.month
df['day'] = df['started_at'].dt.day
df['year'] = df['started_at'].dt.year
```

```python
df.describe()
```

```python
counts = df['year'].value_counts()
print(counts)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df = df.drop(columns=['ride_id'])
```

```python
df = df.drop(columns=['started_at'])
```

```python
df = df.drop(columns=['ended_at'])
```

```python
df.head()
```

```python
df.tail()
```

```python
counts = df['year'].value_counts()
print(counts)
```

```python
# Check for completely duplicate rows
duplicate_rows = df[df.duplicated()]

# Display them
print(duplicate_rows)

```

```python
df[df.duplicated(keep=False)]
```

```python
# Check for missing values in the entire DataFrame
missing_values = df.isnull()

# Display the boolean DataFrame (True for missing values, False for others)
print(missing_values)
```

```python
# Count missing values per column
missing_count_per_column = df.isnull().sum()

# Display the count of missing values
print(missing_count_per_column)
```

```python
# Check for any remaining missing values
df.isna().sum()

# Or total missing
df.isna().sum().sum()
```

```python
# Filter rows for each year
df_2019 = df[df['year'] == 2019]
df_2020 = df[df['year'] == 2020]
```

```python
print("2019 rides:", len(df_2019))
print("2020 rides:", len(df_2020))

df_2019.head()
df_2020.head()
```

```python
# Save to CSV files
df_2019.to_csv("trips_2019_v3.csv", index=False)
df_2020.to_csv("trips_2020_v3.csv", index=False)
```
